# EDA

## Create a scatterplot of Grad.Rate versus Room.Board where the points are colored by the Private column. 

```python
colors = {'Yes': 'Blue', 'No': 'Red'}
plt.figure(figsize=(12, 6))

# Create scatterplot
plt.scatter(df['Room.Board'], df['Grad.Rate'], c=df['Private'].map(colors), s=20)

# Add labels
plt.xlabel('Room Board')
plt.ylabel('Grad Rate')

# Add grid
plt.grid(True)

# Remove borders
plt.box(False)

# Create legend
legend_elements = [
    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='blue', markersize=8, label='Private: Yes'),
    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='red', markersize=8, label='Private: No')
]
plt.legend(handles=legend_elements, loc='center right', frameon=False, bbox_to_anchor=(1.2, 0.5))

# Show the plot
plt.show()
```

## Create a scatterplot of F.Undergrad versus Outstate where the points are colored by the Private column.

```python
colors = {'Yes': 'Plum', 'No': 'Lightgreen'}
plt.figure(figsize=(12, 6))

# Create scatterplot
plt.scatter(df['Outstate'],df['F.Undergrad'], c=df['Private'].map(colors), s=20)

# Add labels
plt.xlabel('Room Board')
plt.ylabel('Grad Rate')

# Add grid
plt.grid(True)

# Remove borders
plt.box(False)

# Create legend
legend_elements = [
    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='Plum', markersize=8, label='Private: Yes'),
    plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='Lightgreen', markersize=8, label='Private: No')
]
plt.legend(handles=legend_elements, loc='center right', frameon=False, bbox_to_anchor=(1.2, 0.5))

# Show the plot
plt.show()
```

## Use sns.FacetGrid to create a stacked histogram showing Out of State Tuition based on the Private column. 

```python
grid = sns.FacetGrid(data=df, hue="Private", height=6, aspect=1.5)

# Define the histogram plot
grid.map(sns.histplot, "Outstate", bins=20, alpha=0.6)


# Set plot title and labels
plt.grid(True)
plt.xlabel("OutState")

sns.despine()

# Set the background color
plt.gca().set_facecolor("Lightblue")

# Display the plot
plt.show()
```

## Use sns.FacetGrid to create a stacked histogram showing Grad.Rate based on the Private column. 

```python
grid = sns.FacetGrid(data=df, hue="Private", height=6, aspect=1.5)

# Define the histogram plot
grid.map(sns.histplot, "Grad.Rate", bins=20, alpha=0.6)

# Set plot title and labels
plt.grid(True)
plt.xlabel("OutState")
plt.xlabel("Grad.Rate")
plt.gca().set_facecolor("Lightblue")

# Display the plot
plt.show()
```

## K Means Model

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

## Fit the model to all the data except for the Private label.

```python
X = df.drop("Private", axis=1)

# Fit the K-Means model to the data
kmeans = KMeans(n_clusters=2)
pred=kmeans.fit(X)
```

## What are the cluster center vectors?

```python
cluster_centers = kmeans.cluster_centers_


print("Cluster Center Vectors:")
for center in cluster_centers:
        print(center)
```

## Create a new column for df called 'Cluster', which is a 1 for a Private school, and a 0 for a public school.

```python
Cluster = df.copy()
Cluster['Cluster'] = np.where(Cluster['Private'] == 'Yes', 1, 0)

Cluster
```

## Create a confusion matrix and classification report to see how well the Kmeans clustering worked without being given any labels.

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

pred = kmeans.labels_
y = Cluster['Cluster']

conf = confusion_matrix(y, pred)
print(conf)
```

```python
report = classification_report(y, pred)
print(report)
```
