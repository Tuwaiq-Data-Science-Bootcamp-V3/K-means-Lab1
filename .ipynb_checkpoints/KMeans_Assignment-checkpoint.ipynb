## Use sns.FacetGrid to create a stacked histogram showing Out of State Tuition based on the Private column. 

```python
sns.histplot(data=df[df['Private']=='No'], x='Outstate', color='blue', alpha=0.5, bins=20 )
sns.histplot(data=df[df['Private']=='Yes'], x='Outstate', color='orange', alpha=0.5, bins=20)
sns.set_style("darkgrid")
sns.set_style("darkgrid")
plt.xlabel('Out of State')
plt.ylabel('Count')
plt.show()
```

## Use sns.FacetGrid to create a stacked histogram showing Grad.Rate based on the Private column. 

```python
sns.histplot(data=df[df['Private']=='No'], x='Grad.Rate', color='blue', alpha=0.5, bins=20)
sns.histplot(data=df[df['Private']=='Yes'], x='Grad.Rate', color='orange', alpha=0.5, bins=20)
plt.xlabel('Grad Rate')
plt.ylabel('Count')
sns.set_style("darkgrid")
plt.show()
```

## Notice how there seems to be a private school with a graduation rate of higher than 100%.What is the name of that school?

```python
GreatestSchool = df[(df['Private'] == 'Yes') & (df['Grad.Rate'] > 100)]
GreatestSchool
```

## Set that school's graduation rate to 100 then re-do the histogram visualization.

```python
df.loc[df['Grad.Rate'] == 100]
sns.histplot(data=df[df['Private']=='No'], x='Grad.Rate', color='blue', alpha=0.5, bins=20)
sns.histplot(data=df[df['Private']=='Yes'], x='Grad.Rate', color='orange', alpha=0.5, bins=20)
plt.xlabel('Graduation Rate')
plt.ylabel('Count')

plt.show()
```

## K Means Model

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

## Fit the model to all the data except for the Private label.

```python
X = df.drop('Private', axis=1)
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
```

## What are the cluster center vectors?

```python
kmeans.cluster_centers_
```

## Create a new column for df called 'Cluster', which is a 1 for a Private school, and a 0 for a public school.

```python
df['Cluster'] = df['Private'].apply(lambda x: 1 if x == 'Yes' else 0)
```

```python
df.head()
```

## Create a confusion matrix and classification report to see how well the Kmeans clustering worked without being given any labels.

```python
from sklearn.metrics import confusion_matrix
y_true = df['Cluster'].values
y_pred = kmeans.labels_
conf_matrix = confusion_matrix(y_true, y_pred)

# Print the confusion matrix
print(conf_matrix)
```

```python
from sklearn.metrics import classification_report
cr = classification_report(y_true, y_pred)
print(cr)
```
